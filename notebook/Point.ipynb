````python
# creer un animation
fig = plt.figure(1)

nvalues = 50
xmin, xmax = 0.0, 10.0
ymin, ymax = 0.0, 1.0

v = 10.0               # vitesse du point 

ax = fig.add_subplot(111, autoscale_on=False, xlim=(xmin, xmax), ylim=(xmin, ymax))
ax.grid()

line, = ax.plot([], [], 'o-', lw=2)

t = np.linspace(xmin, xmax, nvalues)

x =   t 
y =   0.5 * np.ones(nvalues) * (ymax - ymin)

time_text = ax.text(0.05, 0.1, '', transform=ax.transAxes)

def animate(nframes):
    thisx = [x[nframes], x[nframes]]
    thisy = [y[nframes], y[nframes]]
    line.set_data(thisx, thisy)
    time_text.set_text('temps = %.2fs' % (t[nframes]))
    return line, time_text

anim = animation.FuncAnimation(fig, animate, frames=nvalues)
# call our new function to display the animation
display_animation(anim)
````
